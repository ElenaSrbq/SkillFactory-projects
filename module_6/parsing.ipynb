```python
import pandas as pd
import glob
```

```python
extention = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extention))]
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames])
combined_csv.to_csv('parsing.csv', index=False, encoding='utf-8-sig')
```
